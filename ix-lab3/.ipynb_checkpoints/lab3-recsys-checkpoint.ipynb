### 3.2.1 Baseline recommender system

#### Exercise 3.6 Baseline model

```python
def mean(data):
    ratings = data.map(lambda x : x['rating']).collect()
    return np.sum(ratings)/ len(ratings)
```

```python
def user_bias(data, mean, nbUser):
    biases = np.zeros(nbUser)
    counts = np.zeros(nbUser)
    ratings = data.map(lambda x: (x['userId'], x['rating'])).collect()
    for u, r in ratings:
        biases[u-1] += (r - mean)
        counts[u-1] += 1
    mean_bias_overall = np.sum(biases)/np.sum(counts)
    for u in range(nbUser):
        biases[u] = mean_bias_overall if counts[u]==0 else biases[u]/counts[u]
        
    return biases
```

```python
def item_bias(data, mean, user_bias, nbItem):
    biases = np.zeros(nbItem)
    counts = np.zeros(nbItem)
    ratings = data.map(lambda x : (x['userId'], x['movieId'], x['rating'])).collect()
    for u, m, r in ratings:
        index = movieId2index[m]
        biases[index] += r - user_bias[u-1] - mean
        counts[index] += 1
    mean_bias_overall = np.sum(biases)/np.sum(counts)
    for i in range(nbItem):
        biases[i] = mean_bias_overall if counts[i]==0 else biases[i]/counts[i]
    
    return biases
```

#### Exercise 3.7 Evaluation

```python
def evaluation_1(train, test, nbUser, nbItem):
    means = mean(train)
    user_biases = user_bias(train, means, nbUser)
    item_biases = item_bias(train, means, user_biases, nbItem)
    users = np.zeros(nbUser)
    counts = np.zeros(nbUser)
    
    ratings = test.map(lambda x: (x['userId'], x['movieId'], x['rating'])).collect()
    for u, m, r in ratings:
        index = movieId2index[m]
        users[u-1] += (means + user_biases[u-1] + item_biases[index] - r) ** 2
        counts[u-1] += 1
    for u in range(nbUser):
        if counts[u] != 0 :
            users[u] = users[u]/counts[u] 
    users = np.sqrt(users)
    
    return np.sum(users)/np.count_nonzero(users)

err = evaluation_1(train, test, nbUser, nbItem)
print("We have an error {e}".format(e=err))
```

### 3.2.2 Matrix-factorization model

#### Exercise 3.8 Regularization

```python
def regularization(data, rank_, lambda__, nbIter ):
    data = data.map(lambda x : Rating(x['userId'], x['movieId'], x['rating']))
    return ALS.train(data,rank=rank_, lambda_=lambda__ ,iterations=nbIter)
```

```python
def error(model, data):
    to_pred = data.map(lambda x : (x['userId'], x['movieId']))
    pred = model.predictAll(to_pred)
    pred = pred.map(lambda x : ((x[0],x[1]),x[2]))
    true = data.map(lambda x : ((x['userId'],x['movieId']),x['rating']))
    # join
    true = true.join(pred)
    MSE = true.map(lambda x: (x[1][0] - x[1][1])**2).mean()
    RMSE = sqrt(MSE)
    return RMSE
    
```

```python
nbIter = 10
rank = 15
def evaluation_2(train, test):
    lam = [1.e-4, 1.e-3, 1.e-2, 1.e-1]
    for lambda_ in lam:
        model = regularization(train, rank, lambda_, nbIter)
        err = error(model, test)
        print ("With rank = {r} and lambda = {l}, we have an error {e}".format(r=rank,l=lambda_,e=err))

evaluation_2(train, test)  
```

##### Rank = R means that we decompose each film/user in R features. So the trade-off is the multiplication between two matrix Q (shape = (N,R), N : nbUser) and Z (shape = (M,R), M: nbMovie). If we choose a lower rank R then we will have less computation then the one of higher rank.

#### With lambda = 1.e-4, we have a higher error then the baseline method. Because 1.e-4 is actually small and not enough for penalizing the noises in the data. We can see that with lambda = 1.e-3, the situation are improved (error of 0.84)....

#### Exercise 3.10 Visualisation

```python
import pickle
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.io import output_notebook
from collections import OrderedDict
with open("selected-movies.pickle", "rb") as f:
    sel_movs = pickle.load(f, encoding="utf8")
```

```python
def visualization(x_, y_, names_, colors_):
    output_notebook()
    colors = ["#%02x%02x%02x" % (int(r*255),int(r*255),int(r*255)) for r in colors_]
    source = ColumnDataSource(
        data=dict(
                    x=x_,
                    y=y_,
                    label=names_
                 )
    )
    TOOLS="pan,wheel_zoom,box_zoom,reset,hover,previewsave"
    p = figure(title="Tomatoes score for some movies, color is white if score = 1 and black if score = 0", tools=TOOLS)
    p.circle('x', 'y', fill_color=colors, size=20, source=source)

    hover =p.select(dict(type=HoverTool))
    hover.tooltips = OrderedDict([("name", "@label"),])
    show(p)
```

```python
def movie_visualization(train, sel_movs, data_movies):
    model = regularization(train, 2, 1.e-1, 10)
    dic = dict(model.productFeatures().collect())
    features = []
    for (id_, title, score) in sel_movs:
        features.append(dic[id_])
    x = [x[0] for x in features]
    y = [x[1] for x in features]
    color = [x[2] for x in sel_movs]
    movie2genres = []
    for movie in sel_movs:
        genres = data_movies.filter(lambda x : x['movieId']==movie[0]).map(lambda x : x['genres']).collect()
        movie2genres.append((movie[1],genres))
    visualization(x, y, movie2genres, color)
    
movie_visualization(train, sel_movs, data_movies)   
```

#### If the film has a high Tomatoes score than it's placed on x-axis (the more high score the more it goes to the right) and if the film has a low score that it's placed on y-axis (the more low score the more it goes up ).

#### It seems that x-axis represents WAR, THRILLER films, and y-axis represents COMEDY, ROMANCE films
