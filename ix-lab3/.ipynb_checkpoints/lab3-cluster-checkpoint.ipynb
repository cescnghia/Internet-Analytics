### Exercise 3.11 Clustering tags

```python
def kmean_dimension(d):
    one_d   = [d[:1] for d in d.values()]
    two_d   = [d[:2] for d in d.values()]
    three_d = [d[:3] for d in d.values()]
    four_d  = [d[:4] for d in d.values()]
    five_d  = [d[:5] for d in d.values()]
    
    return one_d, two_d, three_d, four_d, five_d

one_d, two_d, three_d, four_d, five_d = kmean_dimension(d)
```

```python
def visualization_1D(data,k):
    kmeans2d = KMeans(n_clusters=k, random_state=0).fit(data)
    x = [x[0] for x in data]
    y = [0*s for s in x]
    output_notebook()
    source = ColumnDataSource(
        data=dict(
                    x=x,
                    y=y,
                    label=list(tags)
                 )
    )
    box_color = ["#0000ff","#ff0066","#996633","#33cc33"]
    colors_ = [box_color[r] for r in kmeans2d.labels_]

    TOOLS="pan,wheel_zoom,box_zoom,reset,hover,previewsave"
    p = figure(title="Visualization of {s}-cluster on one principle direction".format(s=k), tools=TOOLS)
    p.circle('x', 'y', fill_color=colors_, size=10, source=source)

    hover =p.select(dict(type=HoverTool))
    hover.tooltips = OrderedDict([("tag", "@label"),])

    show(p)
visualization_1D(one_d,2)
```

```python
def visualization_2D(data,k):
    kmeans2d = KMeans(n_clusters=k, random_state=0).fit(data)
    x = [x[0] for x in data]
    y = [x[1] for x in data]
    output_notebook()
    source = ColumnDataSource(
        data=dict(
                    x=x,
                    y=y,
                    label=list(tags)
                 )
    )
    box_color = ["#0000ff","#ff0066","#996633","#33cc33"]
    colors_ = [box_color[r] for r in kmeans2d.labels_]

    TOOLS="pan,wheel_zoom,box_zoom,reset,hover,previewsave"
    p = figure(title="Visualization of {s}-cluster on two principle directions".format(s=k), tools=TOOLS)
    p.circle('x', 'y', fill_color=colors_, source=source)

    hover =p.select(dict(type=HoverTool))
    hover.tooltips = OrderedDict([("tag", "@label"),])
    show(p)

visualization_2D(two_d,2)
#visualization_2D(two_d,4)
```

### Exercise 3.12 Clustering movies

```python
with open("most-rated.pickle", "rb") as f:
    most_rated = pickle.load(f, encoding="utf8")
data = sc.textFile("/ix/ml-20m/movies.txt").map(json.loads)
datum = data.collect()
```

```python
def films_to_genres(datum, most_rated):
    dictionary = {}
    ids_genres = datum.map(lambda x : (x['movieId'],x['genres'])).collect()
    for (id_1, film_name) in most_rated:
        for (id_2, genres) in ids_genres:
            if(id_1 == id_2):
                dictionary[film_name] = genres
    return dictionary
```

```python
def jaccard_distance(a,b):
    return 1.0 - len(a.intersection(b))/len(a.union(b))
```

```python
def k_medioids(k, dictionary):
    moviesId = list(dictionary.keys())
    
    #Initialize medioids
    m = random.sample(set(moviesId),k)
    
    newDist = [0]*k
    epsilon = [sys.maxsize]*k
    clusters = [[] for _ in range(k)]
    oldClusters = [[] for _ in range(k)]
    continue_ = True
    
    while(continue_):
        continue_ = False
        
        #assign point to cluster     
        clusters = [[] for _ in range(k)]
        for i in range(len(moviesId)):
            dist = [0]*len(m)
            for j in range(len(m)):
                dist[j] = jaccard_distance(set(dictionary[m[j]]), set(dictionary[moviesId[i]]))
            minDistance = min(dist)
            clusterIndex = dist.index(minDistance) 
            clusters[clusterIndex].append(moviesId[i])
        
        #recompute medioids
        for i in range(k):
            dist = [0]*len(clusters[i])
            for j in range(len(clusters[i])):
                sumDist = 0
                for z in range(len(clusters[i])):
                    a = set(dictionary[clusters[i][j]])
                    b = set(dictionary[clusters[i][z]])
                    sumDist += jaccard_distance(a,b)
                dist[j] = sumDist
            minDist = min(dist)
            # new medioid
            m[i] = clusters[i][dist.index(minDist)]
            newDist[i] = minDist
        
        #check if the cluster stayed the same after the iteration
        for i in range(k):
            epsilon = set(clusters[i]).intersection(set(oldClusters[i]))
            if((len(epsilon) != len(clusters[i])) or (len(epsilon) != len(oldClusters[i]))):
                continue_ = True
                
        oldClusters = clusters
        
    return clusters
```

```python
def movieIdToGender(c, dictionary):
    clusters = [[] for _ in range(len(c))]
    for i in range(len(c)):
        for j in range(len(c[i])):
            for b in range(len(dictionary[c[i][j]])):
                clusters[i].append(dictionary[c[i][j]][b])
    return clusters
```

```python
def visualizeCluster(clusters):
    genders =  list(set(clusters[0]).union(set(clusters[1])))
    genderFreqC1 = [""]*len(genders)
    genderFreqC2 = [""]*len(genders)
    for i in range(len(genders)):
        genderFreqC1[i] = clusters[0].count(genders[i])
        genderFreqC2[i] = clusters[1].count(genders[i])
    mpl_fig = plt.figure()
    ax = mpl_fig.add_subplot(111)
    ind = np.arange(len(genders))
    width = 0.35 

    p1 = ax.bar(ind, genderFreqC1, width, color=(0.2588,0.4433,1.0), label = "cluster 1")
    p2 = ax.bar(ind, genderFreqC2, width, color=(1.0,0.5,0.62), bottom=genderFreqC1, label = "cluster 2")
    ax.set_ylabel('genres frequency')
    ax.set_xlabel('genres')
    ax.set_title('Clustering movies based on genres')

    labels = genders
    ax.xaxis.set_ticks(np.arange(0, 19, 1))
    ax.set_xticklabels(labels = labels,rotation = 90)
    labels = [item.get_text() for item in ax.get_xticklabels()]

    ax.set_xticklabels(labels)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
c = k_medioids(2,films_to_genres(data, most_rated))
visualizeCluster(movieIdToGender(c, films_to_genres(data, most_rated)))
```

####  It depends on the random values chosen for m. If they are close than the cluster separation is not so good, but if they are different for ex. m1 is (Comedy, Romance) and m2 is (Thriller, Sci-Fi) than the clusters are well separated.
