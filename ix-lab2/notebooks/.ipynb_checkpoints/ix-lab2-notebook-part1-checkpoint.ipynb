#### Exercise 2.2:  Node degree distribution

```python
def plot_log_log_distribution(x,y, title, ytitle): 

    plt.subplot(111)
    plt.title(title)
    plt.xlabel("Degree")
    plt.ylabel(ytitle)
    plt.loglog(x, y)
    plt.show()
```

```python
def compute_degrees(G): 
    degs = {}
    # For each node in the graph, we look at its degree and put into the dictionary
    # E.g : degs[3] = {1,6,7} means that nodes 1, 6, 7 have the same degree 3
    for n in G.nodes():
        deg = G.degree(n) 
        if deg not in degs:
            degs[deg] = 0 
        degs[deg] += 1
        
    items = sorted(degs.items())
    
    # x = array of all degrees 
    # y = array of number of nodes with degree that corresponds to array x
    x, y = zip(*items)
    
    
    # z = array of number of node that has degree greater than x
    z = (nb_nodes, )
    current = nb_nodes
    for i in y[:-1]:
        current -= i
        z += (current,)
    
    return x, y, z

x, y, z = compute_degrees(G_1)

plot_log_log_distribution(x,y, "Log-log distribution", "Number of nodes with degree x")
plot_log_log_distribution(x,z, "Log-log cummulative distribution", "Number of nodes with degree >= x")
```

2.2.1 We first draw the log-log plot but it is not so usefull in our case because the plot doesn't really follow the power law (slope -Î³). So we switched to the log-log cumulative plot which follows the power law.

2.2.2 The plot is linear and it follows the power law.

2.2.3 It is close to the Pareto probability distribution with a parameter Î³ = 4\3

#### Exercise 2.4: Short paths & Small-worlds

```python
path_lengths_1 = nx.all_pairs_shortest_path_length(G_1)
```

```python
# (key,value) where key is the path length and value is the number of paths with length equal to key
dict_lengths_1 = {}

for i in G_1.nodes():
    for y in path_lengths_1[i].values():
        if y not in dict_lengths_1:
            dict_lengths_1[y] = 0
        dict_lengths_1[y] += 1  

dict_lengths_1.pop(0, None) 

# Because we count twice the path lenght,so we have to divide by 2
for i in dict_lengths_1.keys():
    dict_lengths_1[i] /= 2
```

```python
def plot_degree_distribution_1(dictionary, title, xtitle, ytitle):
    plt.bar(range(len(dictionary)), list(dictionary.values()), align='center')
    plt.xticks(range(len(dictionary)), list(dictionary.keys()))
    plt.xlabel(xtitle)
    plt.ylabel(ytitle)
    plt.title(title)
    plt.show()
    
```

```python
plot_degree_distribution_1(dict_lengths_1,'Distribution of path lengths' , 'Lenght ', 'Number of pair has distance x ')
```

This is the normal distribution. We need in average 4 hops to reach any other node in the network. So it is a small world.
